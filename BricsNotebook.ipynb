## Problem Definition:
Why is the relationship between Stellar Mass (M*) and Dark Matter Halo Mass (Mh) important in astrophysics?

-**The relationship between a galaxy’s stellar mass (M ) and its dark matter halo mass (Mh) lies at the heart of studies on galaxy formation and evolution. It allows us to** :

- Understand how baryons (i.e., ordinary matter such as gas) are converted into stars, which is crucial for modeling star formation efficiency.

- Investigate galactic dynamics through observable properties such as HI rotation velocity, total mass, luminosity, and stellar population age.

- Highlight discrepancies between theoretical predictions and observations, particularly to identify the influence of stellar feedback, metallicity, or the broader cosmological environment.

#### 1-stellarmass_sparc.txt data :

- **Title** : The Baryonic Tully-Fisher Relation II: Stellar Mass Models
- **Authors** : Duey F, Schombert J, McGaugh S, Lelli F
- **Table** : WISE Stellar Masses
- **Dimension of Sparc stellar Data** : 111 rows × 6 columns
- **Number of Galaxies** : 111
- **Name of the columns** : "Galaxy", "W1_Luminosity", "Stellar_Mass", "e_Stellar_Mass", "Color_g_W1", "M_to_L"
- **Data types** : float64(5), object(1)
- **Number of NaN values** : 0 missing values

- **Data Column Descriptions**:
    - Galaxy : Name of the galaxy.

    - W1_Luminosity (log(L☉)) : Total luminosity in the W1 band (3.4 μm) → traced by WISE photometry.

    - Stellar_Mass (log(M☉)) : Total stellar mass → estimated from W1 and the M/L ratio.

    - e_Stellar_Mass (log(M☉)) : Mean error on stellar mass.
    
    - Color_g_W1 (mag) : g−W1 color = difference between SDSS g (visible) and WISE W1 (near-infrared) magnitudes → indicator of stellar age or metallicity.

    - M_to_L : Stellar mass to light ratio in the W1 band → used to convert light into mass.

#### 2-Halomass_sparc.txt data :

- **Title** : The Halo Mass Function of Late-Type Galaxies from HI Kinematics
- **Authors** : P. Li, F. Lelli, S.S. McGaugh, M.S. Pawlowski, M. Zwaan, J.M. Schombert
- **Dimension of Sparc stellar Data** : 175 rows × 9 columns
- **Number of Galaxies** : 175
- **Name of the columns** : "Galaxy", "WP50", "e_WP50","Halo_Mass_NFW", "e_Halo_Mass_NFW","Halo_Mass_Ein", "e_Halo_Mass_Ein",
"Halo_Mass_DC14", "e_Halo_Mass_DC14"
- **Data types** : float64(8), object(1)
- **Number of NaN values** : 0 missing values

- **Data Column Descriptions**:
    - Galaxy : Name of the galaxy.

    - WP50 (km/s) : HI line width at 50% of peak flux → measure of galaxy rotational velocity.

    - e_WP50 (km/s) : Mean error on WP50.

    - Halo_Mass_NFW (log(M☉)) : Dark matter halo mass based on the NFW model (Navarro-Frenk-White, 1997).
    
    - e_Halo_Mass_NFW (log(M☉)) : Mean error on Halo_Mass_NFW.

    - Halo_Mass_Ein (log(M☉)) :  Halo mass based on the Einasto model (Ludwig Einasto, 1965).

    - e_Halo_Mass_Ein (log(M☉)) : Mean error on Halo_Mass_Ein.

    - Halo_Mass_DC14 (log(M☉)) : Halo mass based on the DC14 model (Di Cintio et al., 2014).

    - e_Halo_Mass_DC14 (log(M☉)) : Mean error on Halo_Mass_DC14.

- **Title** : The extended GALEX Arecibo SDSS Survey (xGASS) Catinella et al. 2018
- **Authors** : Barbara Catinella

    - GASS Data Release 2: Catinella et al 2012, A&A 544, A65

- **Dimension of xGASS Data** : 1178 rows, 42 columns
- **Number of Galaxies** : 1178
- **Name of the columns** : 'GASS', 'AGCnr', 'SDSS', 'PLATEID', 'MJD', 'FIBERID', 'RA', 'DEC',
       'zSDSS', 'Dlum', 'lgMstar', 'expAB_r', 'model_r', 'merr_r', 'ext_r',
       'petrR50_r', 'petrR90_r', 'petrR50_z', 'lgmust', 'CINDX', 'INCL',
       'NUVr', 'lgSFR_tot_median', 'SFR_best', 'SFRerr_best', 'SFRcase_best',
       'HIsrc', 'HI_FLAG', 'xGASSDR', 'zHI', 'W50cor', 'lgMHI', 'lgGF',
       'HIconf_flag', 'weight', 'group_id_B', 'NYU_id', 'env_code_B', 'Ngal_B',
       'yang_shred_code', 'logMh_Mst_B', 'lvir_ratB'
- **Data types** : float64(27), int64(14), object(1)
- **Number of NaN values** : 0 missing values

- **Data Column Descriptions**:
    - GASS : GASS/xGASS identifier (GASS-low if >100000)
    - AGCnr : AGC identifier (if available)
    - SDSS : SDSS identifier
    - PLATEID : SDSS spectroscopic Plate ID
    - MJD : SDSS spectroscopic MJD
    - FIBERID : SDSS spectroscopic Fiber ID
    - RA : Right Ascension (J2000) from SDSS
    - DEC : Declination (J2000) from SDSS
    - zSDSS	: SDSS spectroscopic redshift
    - Dlum	: luminosity distance in Mpc (assuming Omega_m=0.3, Omega_L=0.7, H0=70 km/s/Mpc)
    - lgMstar : base-10 logarithm of stellar mass, in solar units, from SDSS MPA/JHU
    - expAB_r : minor-to-major axial ratio from the exponential fit in r band, from SDSS
    - model_r : r band model magnitude from SDSS
    - merr_r : error on r band model magnitude from SDSS
    - ext_r  : Galactic extinction in r band, in magnitudes
    - petrR50_r  : radius containing 50% of the Petrosian flux in r band, in arcsec
    - petrR90_r  : radius containing 90% of the Petrosian flux in r band, in arcsec
    - petrR50_z  : radius containing 50% of the Petrosian flux in z band, in arcsec
    - lgmust     : base-10 logarithm of stellar mass surface density, in Msun/kpc^2
    - CINDX      : concentration index in r band (petrR90_r/ps.petrR50_r).
    - INCL       : inclination to the line-of-sight, in degrees, computed from the above
	      	    axis ratio (see GASS DR2 paper for details).
    - NUVr	: NUV-r colour, corrected for Galactic extinction, in mag
    - lgSFR_tot_median  : base-10 logarithm of total SFR, Msun/yr, from SDSS MPA/JHU
    - SFR_best  : Final best SFR from UV+MIR    [Msun/yr]
    - SFRerr_best : uncertainty on SFR_best       [Msun/yr]
    - SFRcase_best : case of UV+MIR SFR calculation
	       	    1- GALEX NUV + WISE W4 (both unflagged)
	       	    2- GALEX NUV + WISE W3 (both unflagged)
	       	    6- GALEX NUV or WISE W3 or W4
	       	    9- SED SFRs corrected to match UV+MIR (no error estimates)
    - HIsrc  : source of HI data:
	     	    1- ALFALFA 70% (web release: https://egg.astro.cornell.edu/alfalfa/data/)
	      	    2- Cornell HI archive (Springob et al. 2005, ApJS, 160, 149)
	       	    3- xGASS detection
	      	    4- xGASS non-detection
    - HI_FLAG  : quality flag for xGASS detections (1=good, 2=marginal, 5=confused). We
	      	    assigned code 3 only to a few galaxies, which are both marginal and confused.
    - xGASSDR : GASS Data Release (1, 2 or 3) or GASS-low (4)
    - zHI : redshift measured from the HI spectrum.
    - W50cor  : velocity width corrected for instrumental broadening and cosmological redshift
	      	    only, in km/s. No inclination or turbulent motion corrections are applied.
    - lgMHI  : base-10 logarithm of the HI mass, in solar units. This is an upper limit
	      	    if HIsrc=4 (computed assuming a 5 sigma signal with a 200/300 km/s velocity
	      	    width for GASS-low/GASS respectively, if the spectrum was smoothed to half width)
    - lgGF  : base-10 logarithm of the HI mass fraction, MHI/Mst. For non-detections, this
	      	    is computed using the upper limit on the HI mass
    - HIconf_flag : HI confusion flag (1= confusion is certain; >0= small companion; 0= not confused).
              	    Set to -99 for non-detections (HIsrc=4, for which confusion was not assessed).
    - weight : weights based on the GAMA stellar mass function [Baldry et al. 2012]
	      	    to recover the stellar mass distribution of a volume-limited sample
    - group_id_B  : group ID
    - NYU_id   : NYU identifier
    - env_code_B  : environment identity:
                    0: satellite
                    1: isolated central
                    2: group central
                   -1: not in group catalog
    - Ngal_B  : number of galaxies in group (0 if not in group catalog)
    - yang_shred_code : whether galaxy was corrected for shredding in group B(II)
                    (0= no, 1= yes, correction was required)
    - logMh_Mst_B  : halo mass (using Mst) (0 if not in group catalog)
    - lvir_ratB  : log10 r/r_vir, ratio of distance to group cent over group virial radius

In this plot:

```python
# Plot2 : Simplified Theoretical Model (Add a theoretical curve (Behroozi-like model))
def behroozi(logMh):
    return 10.8 + 1.5*(logMh - 12) - 0.5*(logMh - 12)**2 # Define the Behroozi-like model

x = np.linspace(9, 13, 100)
y = behroozi(x)

plt.figure(figsize=(10,6))
plt.scatter(sparc_combined["Halo_Mass_NFW"], sparc_combined["Stellar_Mass"], label="NFW", color='blue', alpha=0.7)
plt.scatter(sparc_combined["Halo_Mass_Ein"], sparc_combined["Stellar_Mass"], label="Einasto", color='green', alpha=0.7)
plt.scatter(sparc_combined["Halo_Mass_DC14"], sparc_combined["Stellar_Mass"], label="DC14", color='purple', alpha=0.7)

plt.plot(x, y, 'r--', lw=2, label="Behroozi Model (2013)")
plt.xlabel("Halo Mass")
plt.ylabel("Stellar Mass")
plt.title("Stellar–Halo Mass Relation (SPARC)")
plt.legend()
plt.grid(True)
plt.show()
```

Plot 2: We have overlaid a theoretical relation of the Behroozi (2013) type onto the SPARC data.

-A clear positive correlation exists between stellar mass and halo mass: the higher the stellar mass of a galaxy, the more massive its halo tends to be.

-The Behroozi (2013) curve is globally followed , but discrepancies are present, especially for galaxies with extreme stellar masses.

-This shows that cosmological predictions are broadly valid , but need to be refined with local astrophysical effects (such as feedback and environment).
